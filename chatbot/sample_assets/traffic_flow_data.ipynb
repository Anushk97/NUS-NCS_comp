```python
import pandas as pd
```

```python
df = pd.read_json('./trafficflow.json')
df['Value'][0]
```

```python
def adjust_date(info):
    date_parts = info['Date'].split('/')
    if date_parts[2] == '2023' and date_parts[1] == '11':
        # Change year from 2023 to 2024 and month from 11 to 03
        info['Date'] = f'{date_parts[0]}/03/2024'
    return info

df['Value'] = df['Value'].apply(adjust_date)
```

```python
df['Value'][:10]
```

```python
df = df.join(pd.json_normalize(df['Value'])).drop('Value', axis=1)
df
```

```python
df.to_json('./trafficflow.json', orient='records')
```

```python
df.to_csv('traffic_flow.csv')
```
